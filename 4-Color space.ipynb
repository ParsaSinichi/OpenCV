```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Split the color channels 

```python
img=cv2.imread("./images/color.png")
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
b,g,r=cv2.split(img)# split the image into B,G,R components
plt.figure(figsize=[20,5])
#For the red channel 
plt.subplot(141)
plt.imshow(r,cmap='gray')
plt.title("Red Channel")
#For the green channel 
plt.subplot(142)
plt.imshow(g,cmap='gray')
plt.title("Green Channel")
#For the blue  channel 
plt.subplot(143)
plt.imshow(b,cmap='gray')
plt.title("Blue Channel")

# Show the original image 
plt.subplot(144)
plt.imshow(img_rgb)
plt.title("All  Channels")
```

Note : here white means we have the channel 

```python
img=cv2.imread("./images/Cow.jpg")
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
b,g,r=cv2.split(img)# split the image into B,G,R components
plt.figure(figsize=[20,5])
#For the red channel 
plt.subplot(141)
plt.imshow(r,cmap='gray')
plt.title("Red Channel")
#For the green channel 
plt.subplot(142)
plt.imshow(g,cmap='gray')
plt.title("Green Channel")
#For the blue  channel 
plt.subplot(143)
plt.imshow(b,cmap='gray')
plt.title("Blue Channel")

# Show the original image 
plt.subplot(144)
plt.imshow(img_rgb)
plt.title("All  Channels")
```

 # Merge the collor channels 

```python
merged=cv2.merge([b,g,r])
plt.figure(figsize=[20,5])
#For the red channel 
plt.subplot(141)
plt.imshow(r,cmap='gray')
plt.title("Red Channel")
#For the green channel 
plt.subplot(142)
plt.imshow(g,cmap='gray')
plt.title("Green Channel")
#For the blue  channel 
plt.subplot(143)
plt.imshow(b,cmap='gray')
plt.title("Blue Channel")

# Show the original image 
plt.subplot(144)
plt.imshow(merged[:,:,::-1])
plt.title("Merged")
```

```python
zeros=np.zeros_like(r,'uint8')
red=cv2.merge([zeros,zeros,r])
green=cv2.merge([zeros,g,zeros])
blue=cv2.merge([b,zeros,zeros])

# Show the channels
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(red[...,::-1],cmap='gray');plt.title("Red");
plt.subplot(142);plt.imshow(green[...,::-1],cmap='gray');plt.title("Green");
plt.subplot(143);plt.imshow(blue[...,::-1],cmap='gray');plt.title("Blue");
```

# HSV Color Channel 

Hue – Color Value (0 – 179)
Saturation – Vibrancy of color (0-255)
Value – Brightness or intensity (0-255)

```python
img=cv2.imread('./images/Cow.jpg')#displaying seprated HSV channels 
hsv_img=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow("HSV image",hsv_img)
cv2.imshow("Hue channel",hsv_img[:,:,0])
cv2.imshow("Saturation channel",hsv_img[:,:,1])
cv2.imshow("Value channel",hsv_img[:,:,2])
cv2.waitKey()
cv2.destroyAllWindows()
```

```python

saturated_image = cv2.merge([hsv_img[:, :, 0], cv2.add(hsv_img[:, :, 1], 50), hsv_img[:, :, 2]])
saturated_image = cv2.cvtColor(saturated_image, cv2.COLOR_HSV2BGR)

plt.figure(figsize=[12,5])

plt.subplot(121)
plt.imshow(saturated_image[...,::-1])
plt.title('saturated image')

plt.subplot(122)
plt.imshow(img[...,::-1])
plt.title('original image')
                            
```

```python
brighter_image = cv2.merge([hsv_img[:, :, 0], hsv_img[:, :, 1], cv2.add(hsv_img[:, :, 2], 50)])
brighter_image = cv2.cvtColor(brighter_image, cv2.COLOR_HSV2BGR)

plt.figure(figsize=[12,5])

plt.subplot(121)
plt.imshow(brighter_image[...,::-1])
plt.title('brighter image')

plt.subplot(122)
plt.imshow(img[...,::-1])
plt.title('original image')
```
