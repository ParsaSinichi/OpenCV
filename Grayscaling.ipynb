# Grayscale images with OpenCV

```python
img=cv2.imread('./images/Cow.jpg')#load our image
cv2.imshow('Original',img)
cv2.waitKey()

#Convert the image to Grayscale
gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(type(gray_image))
cv2.imshow("Grayscale",gray_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Loading a Grayscale image in a faster method 
img=cv2.imread('./images/Cow.jpg',0)# here 0 is a shortcut for Grayscale
print(img.shape)# in Grayscale  we don't have color channels 
cv2.imshow("Gray",img)
cv2.waitKey()
cv2.destroyAllWindows()
```
