# Changing and replacing pixels

```python
img=cv2.imread('./images/Cow.jpeg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img[200:300,200:800,:]=0#> 255 =white, 0=Black

plt.imshow(img)
```

```python
img=cv2.imread('./images/Cow.jpeg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img[200:300,200:800,:]=(255,255,0) 
plt.imshow(img)
```
