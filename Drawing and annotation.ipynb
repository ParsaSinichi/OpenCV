```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create a black image
image=np.zeros((512,512,3),np.uint8)

image_bw=np.zeros((512,512),np.uint8)

plt.imshow(image[...,::-1])
plt.figure()
plt.imshow(image_bw,cmap='gray')
```

    cv2.line(img,starting cordinates,ending cordinates,color,thickness) => for drawing a line 

```python
#Draw a blue line 
img=np.zeros((512,512,3),np.uint8)
cv2.line(img,(0,0),(511,511),(255,127,0),5)#BGR Channel
cv2.imshow("Blue line ",img)
plt.imshow(img[...,::-1])
cv2.waitKey(0)
cv2.destroyAllWindows()

```

# Drawing a rectangle

cv2.rectangle(img,starting vortex,opposite vortex,color, thickness)

```python
img=np.zeros((512,512,3),np.uint8)
cv2.rectangle(img,(100,100),(300,250),(0,0,255),5)
cv2.imshow('rectangle',img)
plt.imshow(img[...,::-1])
cv2.waitKey(0)
cv2.destroyAllWindows()
    
```

```python
img=cv2.imread('images/Cow.jpg')
cv2.rectangle(img,(250,200),(850,550),(0,255,255),(5))
plt.imshow(img[...,::-1])
```
