```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Making a square 
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -1)# here we are using Grayscale so we have 2dimensions
#Making a ellipse 
ellipse = np.zeros((300, 300), np.uint8)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)
#Plot
plt.subplot(121);plt.imshow(square,cmap='gray');plt.title("square");
plt.subplot(122);plt.imshow(ellipse,cmap='gray');plt.title("ellipse");
```

Bitwise operations in opencv : 
Or -> bitwise_or
and -> bitwise_and

```python
#The operands should be in the same dimension
bitwise_and=cv2.bitwise_and(square,ellipse)# AND

bitwise_or=cv2.bitwise_or(square,ellipse)# OR

bitwise_xor=cv2.bitwise_xor(square,ellipse)#XOR

bitwise_not=cv2.bitwise_not(square)#Not

#Plot
plt.figure(figsize=[12,6])
plt.subplot(242);plt.imshow(square,cmap='gray');plt.title("Square");
plt.subplot(243);plt.imshow(ellipse,cmap='gray');plt.title("Ellipse");
plt.subplot(245);plt.imshow(bitwise_and,cmap='gray');plt.title("And");
plt.subplot(246);plt.imshow(bitwise_or,cmap='gray');plt.title("Or");
plt.subplot(247);plt.imshow(bitwise_xor,cmap='gray');plt.title("Xor");
plt.subplot(248);plt.imshow(bitwise_not,cmap='gray');plt.title("not (square)");
#white =1 , black =0 
```

```python
img=cv2.imread('./images/Cow.jpg')
cropped=img[100:600,150:650]
circle=np.zeros((500,500,3),np.uint8)# here we have 3 color channel
cv2.circle(circle,(250,250),250,(255,255,255),-1)
output_img=cv2.bitwise_and(cropped,circle)
plt.imshow(output_img[...,::-1])#BGR to RGB
```

```python
png_image = np.zeros((500,500,4), np.uint8)#4 color channel 
png_image[:,:,0:3] = cropped
circle = np.zeros((500,500), np.uint8)# one color channel
cv2.circle(circle, (250, 250), 250, (255,255,255), -1) #White circle
png_image[:,:,3]= circle
cv2.imwrite("cow.png", png_image)
```

```python
square = np.zeros((10, 10), np.uint8)
cv2.rectangle(square, (2, 2), (8, 8), 2, -1) # 2 : 010

circle1 = np.zeros((10, 10), np.uint8)
cv2.circle(circle1, (5, 5), 2, 3, -1)  # 3 : 011


circle2 = np.zeros((10, 10), np.uint8)
cv2.circle(circle2, (5, 5), 2, 4, -1) # 4 : 100


output1 = cv2.bitwise_and(square, circle1)
output2 = cv2.bitwise_and(square, circle2)

print("--square--")
print(square)
print("--circle1--")
print(circle1)
print("--circle2--")
print(circle2)
print("--output1--")
print(output1)
print("--output2--")
print(output2)
```
