```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#load the image in grayscale
img=cv2.imread('images/notes.png',0)
negative_img=cv2.bitwise_not(img)#inversing the image 
binary_img=cv2.adaptiveThreshold(negative_img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,15,-2)

plt.figure(figsize=[17,6])
plt.subplot(311);plt.imshow(img, cmap='gray');plt.title("Original image in gray");
plt.subplot(312);plt.imshow(negative_img, cmap='gray');plt.title("Negative image");
plt.subplot(313);plt.imshow(binary_img, cmap='gray');plt.title("Binary image");
```

# Extracting horizontal lines 

```python
#creating a structure element 
cols=binary_img.shape[1]
horizontal_size=cols//30 
horizontal_structure = cv2.getStructuringElement(cv2.MORPH_RECT,(horizontal_size,1))
print(horizontal_structure)
```

```python
# Apply morphology operations
horizontal=cv2.morphologyEx(binary_img,cv2.MORPH_OPEN,horizontal_structure)
plt.figure(figsize=[17,6])
plt.subplot(211);plt.imshow(img, cmap='gray');plt.title("Original image in gray");
plt.subplot(212);plt.imshow(horizontal, cmap='gray');plt.title("horizontal lines in image");
```
